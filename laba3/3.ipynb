### Виконання роботи



```python
# Встановлення бібліотеки NumPy
%pip install numpy

# Функція для повідомлень
def homework(message):
    print("\n>>>>>", message, "\n")

# Імпортуємо бібліотеку NumPy
import numpy as np

# Початковий список температур
temperature = [20.1, 20.1, 20.2, 20.4]
print(f"Початковий список температур: {temperature}")

# Спроба множення списку на 5
t1 = temperature * 5
print(f"При множенні на 5 маємо: {t1}")
homework("Цей результат з'явився тому, що у Python операція `список * n` дублює вміст списку n разів, а не множить його елементи.")

# Використання спискового включення для правильного множення
t1 = [x * 5 for x in temperature]
print(f"Тепер отримаємо: {t1}")
homework("Цей результат з'явився тому, що ми використали спискове включення (list comprehension), яке множить кожен елемент списку на 5, створюючи новий список.")

# Створення масиву NumPy
t2 = np.array(temperature)
print(f"Масив з використанням методів NumPy: {t2}")

# Множення масиву NumPy на 5
t3 = t2 * 5
print(f"При множенні на 5 отримуємо: {t3}")
homework("NumPy дозволяє виконувати елементні операції напряму. Операція `t2 * 5` множить кожен елемент масиву на 5, що є частиною оптимізованої роботи з масивами у цій бібліотеці.")
```

# Метод linspace - генерує задану кількістю чисел у певному діапазоні, 50 чисел за замовчуванням


```python
import numpy as np

def homework(message):
    print("\n>>>>>", message, "\n")

# Початкові значення
start = 5
size_of_vec = 100

# Генерація чисел за допомогою np.linspace
a = np.linspace(start, size_of_vec)
print(f"Згенерували {len(a)} чисел в діапазоні {a[0]} до {a[-1]}")
homework("Виведіть кожне друге число згенерованого діапазону. Тут буде потрібен цикл :)")
print("Пояснення: Щоб вивести кожне друге число, використаємо цикл з кроком 2.")
# Виведення кожного другого числа
for i in range(0, len(a), 2):
    print(a[i])

# Зменшення кількості чисел до 5
print("\nМожна зменшити кількість чисел до 5:")
a = np.linspace(start, size_of_vec, 5)
print(f"Результат: {a}")
homework("Збільшіть кількість чисел до 100. Виведіть перших та останніх 10.")
print("Пояснення: Якщо ми задаємо 5 чисел, то вони рівномірно розподіляються між стартовим і кінцевим значенням.")

# Збільшення кількості чисел до 100
a = np.linspace(start, size_of_vec, 100)
print(f"Перші 10 чисел: {a[:10]}")
print(f"Останні 10 чисел: {a[-10:]}")
homework("Збільшіть кількість чисел до 100. Виведіть перших та останніх 10.")
print("Пояснення: Кількість елементів в масиві збільшено до 100, отже між значеннями буде менший крок.")

# Виключення останнього значення з діапазону
print("""Як бачимо, перше та останнє значення завжди входять у згенерований діапазон.
Відповідно при цьому вираховується крок.
Це можна змінити, виключивши останнє значення з визначення діапазону.
""")
print(np.linspace(start, size_of_vec, 5, endpoint=False))
homework("Спробуйте самі порахувати крок.")
print("Пояснення: Параметр 'endpoint=False' виключає останнє значення з діапазону, тому ми отримуємо рівномірний розподіл без останнього числа.")

# Отримання кроку з функцією retstep
_, sp_1 = np.linspace(start, size_of_vec, 5, endpoint=True, retstep=True)
_, sp_2 = np.linspace(start, size_of_vec, 5, endpoint=False, retstep=True)
print(f"Для порівняння, ми можемо визначити крок автоматично, у нашому випадку це {sp_1} та {sp_2}")
homework("Чому маємо такий крок?")
print("Пояснення: Функція 'retstep=True' повертає значення кроку. Перший крок включає останнє значення, а другий — ні, що і впливає на результат.")
```

# Типи масивів/векторів

```python
import numpy as np

def homework(message):
    print("\n>>>>>", message, "\n")

# Тести з новими масивами
new_examples = {
    'Одновимірний вектор': np.array([1, 2, 3, 4, 5]), 
    'Двовимірна матриця': np.array([[1, 2], [3, 4], [5, 6]]), 
    'Трьохвимірна матриця': np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
}

for key in new_examples:
    print(f"Розглядаємо приклад з {key}. \nЗначення x буде: \n", new_examples[key])
    print(f"Тип {key} x: ", type(new_examples[key]))
    print(f"Розмірність/вимір {key} x:", np.ndim(new_examples[key]))
    print(f"Форма {key} x: ", new_examples[key].shape)
    print(f"Кількість елементів х: {np.size(new_examples[key])}")
    print("-=-"*10)

# Пояснення, чому значення для np.ndim та shape різні
homework("""Як Ви думаєте чому маємо такий результат для функції np.ndim (Розмірність/вимір/dimension).
Чому значення для векторів та матриць однакове та не співпдає з значенням shape?""")

# Пояснення, яке буде виведено:
print("""
Розмірність (np.ndim) визначає кількість осей або вимірів у масиві. Це може бути:
- 1: Одновимірний масив (вектор),
- 2: Двовимірний масив (матриця),
- 3: Тривимірний масив і так далі.

Форма (shape) визначає кількість елементів вздовж кожного виміру. Наприклад:
- Для одновимірного масиву shape = (3,) означає 3 елементи в одному вимірі.
- Для двовимірного масиву shape = (3, 1) означає 3 рядки, по 1 елементу в кожному.

Таким чином:
- Вектор-рядок (np.array([1, 2, 3])) має розмірність 1 (одновимірний масив), але форма цього масиву — (3,).
- Вектор-стовпець (np.array([[1], [2], [3]])) має розмірність 2 (двовимірний масив), і форма його — (3, 1).

Значення для np.ndim відрізняється від shape, оскільки форма описує кількість елементів по кожному виміру, а np.ndim дає загальну кількість вимірів. Тому навіть вектор-стовпець може мати розмірність 2, хоча він є "вектором" (одним стовпцем).
""")
```

# Багатовимірні масиви, розуміння вимірів

```python
import numpy as np

examples = {
    '5 рядків': np.array([[1], [2], [3], [4], [5]]), 
    '5 стовпців': np.array([[1, 2, 3, 4, 5]]),
    '5 елементів': np.array([[[1, 2, 3, 4, 5]]])
}

for a in examples:
    print(f"Для {a} маємо shape: \n{examples[a].shape} \nта вигляд: \n{examples[a]}")
    print(f"Розмірність (np.ndim) для {a}: {np.ndim(examples[a])}")
    print("-=-" * 10)

homework(f"Створіть власні масиви з різною кількістю елементів та використайте метод shape для визначення їх вимірів.")

# Пояснення, чому маємо такі результати:
print("""
1. Масив '5 рядків' має shape = (5, 1), що означає 5 рядків і 1 стовпець. Це є двовимірний масив (матриця) з 5 рядками.
2. Масив '5 стовпців' має shape = (1, 5), що означає 1 рядок і 5 стовпців. Це також двовимірний масив, але лише один рядок з 5 елементами.
3. Масив '5 елементів' має shape = (1, 1, 5), що є тривимірним масивом, де є один набір, один рядок та 5 елементів всередині.

Функція shape допомагає визначити кількість вимірів і розміри кожного виміру масиву:
- Для одновимірного масиву (вектора) shape = (N, ), де N — кількість елементів.
- Для двовимірного масиву (матриці) shape = (M, N), де M — кількість рядків, а N — кількість стовпців.
- Для тривимірного масиву shape = (P, M, N), де P — кількість "глибин" масиву, M — кількість рядків, а N — кількість стовпців.

Тепер давайте створимо ще декілька масивів з різною кількістю елементів:
""")

# Приклад додаткових масивів:
example1 = np.array([1, 2, 3, 4, 5])
example2 = np.array([[1, 2], [3, 4], [5, 6]])
example3 = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

# Виведемо їх форми та типи:
print(f"Масив example1 має shape = {example1.shape} та вигляд:\n{example1}")
print(f"Масив example2 має shape = {example2.shape} та вигляд:\n{example2}")
print(f"Масив example3 має shape = {example3.shape} та вигляд:\n{example3}")
```

```python
import numpy as np

# Масив від 0 до 49
A = np.arange(50)

# Створення умови для парних чисел
all_even = (A % 2 == 0)  # парні числа

# Створення комплексної умови для чисел, менших за 30 та поділених на 5 менших за 2
complex = (A < 30) & (A / 5 < 2) 

# Виведення результатів
print(f"""Спочатку потрібно створити умову за якою здійснювати вибірку.
Результат умови буде або {all_even[0]} для {A[0]} або {all_even[1]} для {A[1]}.
Після підставляння умови пошуку парних чисел в масив отримаємо:
{A[all_even]}
Або застосувавти більш комплексну умову з двох порівнянь:
{A[complex]}
""")

# Пояснення:
print("""
1. Умова для парних чисел (A % 2 == 0) дає масив True/False, в якому True вказує на парні числа в масиві.
   - all_even - масив, що містить True для кожного парного числа.

2. Комплексна умова (A < 30) & (A / 5 < 2) вказує, що вибираються числа менше 30, при цьому результат ділення на 5 повинен бути меншим за 2.
   - complex - масив, що містить True для чисел, які задовольняють обидві умови.

3. Для того, щоб отримати конкретні елементи з масиву, ми використовуємо умовну індексацію:
   - A[all_even] повертає всі парні числа з масиву.
   - A[complex] повертає всі числа, які відповідають комплексній умові.

""")

```

# # Восьме завдання

```python
import numpy as np

# Створення векторів
vector1 = np.array([1, 2, 3])
vector2 = np.array([4, 5, 6])

# Створення матриць
matrix1 = np.array([[1, 2], [3, 4], [5, 6]])
matrix2 = np.array([[7, 8], [9, 10], [11, 12]])

# Виведення розмірностей
print(f"Розмірність vector1: {vector1.shape}")
print(f"Розмірність vector2: {vector2.shape}")
print(f"Розмірність matrix1: {matrix1.shape}")
print(f"Розмірність matrix2: {matrix2.shape}")

# Арифметичні операції

# Додавання векторів
vector_sum = vector1 + vector2
print(f"Сума векторів: {vector_sum}")

# Віднімання векторів
vector_diff = vector1 - vector2
print(f"Різниця векторів: {vector_diff}")

# Множення вектора на число
vector_scalar_mult = vector1 * 2
print(f"Множення вектора на скаляр: {vector_scalar_mult}")

# Додавання матриць
matrix_sum = matrix1 + matrix2
print(f"Сума матриць: \n{matrix_sum}")

# Множення матриць
matrix_mult = np.dot(matrix1, matrix2.T)  # Перемноження матриць
print(f"Множення матриць: \n{matrix_mult}")

# Складання вектора і матриці (не обов'язково однакових розмірностей)
try:
    vector_matrix_sum = matrix1 + vector1  # Не можна додавати матрицю і вектор без відповідних розмірностей
    print(f"Сума матриці і вектора: \n{vector_matrix_sum}")
except ValueError as e:
    print(f"Помилка при додаванні вектора і матриці: {e}")
```
